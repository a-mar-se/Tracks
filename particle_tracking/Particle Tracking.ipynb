### Algoritmo de detección

Ahora tenemos que crear una función que tome un fotograma, opere en el y detecte los centroides de cada bola. Esta función devolverá un DataFrame con la estructura explicada arriba.

```python
def detectCirclesImage(img, frame_number=0, display_intermediate_steps=True):
    
    if display_intermediate_steps==True:
        plt.imshow(img, cmap='gray')
        plt.show()
    # Operations are applied to the image
    
    # ---------------- Fran's Code ---------------------
    # Grayscale and Gaussian Blur
    try:
        bw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        bw = cv2.GaussianBlur(bw, (5,5), cv2.BORDER_DEFAULT)
    except:
        bw=img
    # apply grad morph so that I get thick borders. use a circle kernel
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
    grad = cv2.morphologyEx(bw, cv2.MORPH_GRADIENT, kernel)
    
    # binarize intelligently, by combining Otus's and simple binarization
    _, binary_bw = cv2.threshold(grad, 0.0, 255.0, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    
    # apply open morph over a straight line kernel, 
    #so that I redraw in white regions inside thick borders, but only those containing straight lines
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    connected = cv2.morphologyEx(binary_bw, cv2.MORPH_CLOSE, kernel)
    
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
    erosion = cv2.erode(binary_bw, kernel ,iterations = 2)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    dilation =  cv2.dilate(erosion,  kernel)
    
    difference = cv2.absdiff(connected,dilation)
    # add whitened thick borders to b/w image so that I erase only text/symbols with straigth lines
    added = cv2.add(difference,bw)
    # -------------------------------------------------
    
    if display_intermediate_steps==True:
        plt.imshow(added, cmap='gray')
        plt.show()
        
    # Now I binarize the image to get only b/w, the erosion so blobs are not
    # connected and can later be detected as separate contours
    binarized = cv2.threshold(added, 30.0, 255.0, cv2.THRESH_BINARY)
    if display_intermediate_steps==True:
        plt.imshow(binarized[1], cmap='gray')
        plt.show()
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (29, 29))
    eroded = cv2.erode(binarized[1], kernel, iterations = 1)
    if display_intermediate_steps==True:
        plt.imshow(eroded, cmap='gray')
        plt.show() 
        
    # Find contours in the binary image:
    contours, hierarchy = cv2.findContours(eroded,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    # Detect the centroid of each contour and store it on a list
    centroids = []
    for c in contours:
       # calculate moments for each contour
       M = cv2.moments(c)     
       # calculate x,y coordinate of center
       if M["m00"] != 0:
           cX = M["m10"] / M["m00"]
           cY = M["m01"] / M["m00"]
           centroids.append([cX, cY])
       else:
           cX, cY = 0, 0
    
    # Now we get those circles' positions and store them in an appropiate format
    centroids = np.array(centroids)
    N = centroids.shape[0]

    # We first prepare the structure to store detected circles in a format that
    # trackpy can easily work with (for the linking)   
    A = np.zeros((N, 2), dtype=np.float64)
    B = np.full((N, 1), frame_number, dtype=np.int64)
    names = ('x', 'y')
    A = pd.DataFrame(A, columns=names)
    B = pd.DataFrame(B, columns=('frame',))
    circles_tp = pd.concat((A, B), axis=1)
    # Fill that structure:
    circles_tp.x, circles_tp.y = centroids[:,0], centroids[:,1]
    
    
    
    for index, c in circles_tp.iterrows():
        # Draw the outer circle [(x,y), radius, rgb]
        cv2.circle(img, (int(c['x']),int(c['y'])), 28, (127, 0, 255), 2)
    if display_intermediate_steps==True:
        plt.imshow(img, cmap='gray')
        plt.show() 
    
    return circles_tp
```

Podemos llamar a esta función pasando como argumento la primera imagen del video para mostrar cómo se va modificando la imagen:

```python
video = cv2.VideoCapture(videoPath)
n = 0 # Simple acumulador, para llevar la cuenta de por cual frame voy
while(video.isOpened()):
    # Leemos el frame actual y lo asignamos a la variable frame
    frameExists, frame = video.read()
    
    if frameExists==True:
        if n==1:
            circles = detectCirclesImage(frame)
        n+=1
    else:
        break 
print(circles[:15])
```

### Bucle para procesar todo el video

Tras esto solo tendríamos que aplicar la función anterior a todos los fotogramas del video y guardar los datos en un archivo.

```python
A = np.zeros((1, 2), dtype=np.float64)
B = np.zeros((1, 1), dtype=np.int64)
names = ('x', 'y')
A = pd.DataFrame(A, index=('-1',), columns=names)
B = pd.DataFrame(B, index=('-1',), columns=('frame',))
circles_tp = pd.concat((A, B), axis=1)


video = cv2.VideoCapture(videoPath)
n = 0 # Simple acumulador, para llevar la cuenta de por cual frame voy
while(video.isOpened()):
    # Leemos el frame actual y lo asignamos a la variable frame
    frameExists, frame = video.read()
    
    if frameExists==True:
        new_circles = detectCirclesImage(frame, frame_number=n, display_intermediate_steps=False)
        circles_tp = pd.concat((circles_tp, new_circles), axis=0)
        n+=1
    else:
        break  
    
# We delete the first row of circles_tp, since it was only used for 
# initialization and is no longer needed.
circles_tp = circles_tp.drop('-1')
#TODO: Reniciar indexes
circles_tp = circles_tp.reset_index(drop=True)
    
video.release()
print(circles_tp)
```

## Cálculo de velocidades

Con todo esto ya podemos calcular las velocidades de cada partícula. En este caso usaremos la función np.gradient (*second order accurate central differences*). La función para generar un nuevo DataFrame en el que se incluyan las velocidades es:

```python
traj = pd.read_pickle('trajectories.pkl')

def findVelocities(trajectories):
    """ This function admits a 'trajectories' dataframe as an input
        (usually from trackpy) and returns another dataframe where
        instant velocities are included """
        
    col_names = ['vx', 'vy', 'x', 'y', 'frame', 'particle']
    # Creating an empty dataframe to store results
    data = pd.DataFrame(np.zeros(shape=(1, 6), dtype=np.int64), columns=col_names)
    
    for item in set(trajectories.particle):      
        sub = trajectories[trajectories.particle==item]
        
        if sub.shape[0]<=2:
            #Para obviar los casos en los que solo hay una o dos filas de datos
            pass
        else:    
            # print('Deriving velocities for particle:', str(item))
            dvx = pd.DataFrame(np.gradient(sub.x), columns=['vx',])
            dvy = pd.DataFrame(np.gradient(sub.y), columns=['vy',])
        
            new_df = pd.concat((dvx, dvy, sub.x.reset_index(drop=True), sub.y.reset_index(drop=True),
                                sub.frame.reset_index(drop=True), sub.particle.reset_index(drop=True)),
                                axis=1, names=col_names, sort=False)
            data = pd.concat((data, new_df), axis=0)
        
    # This is to get rid of the first 'np.zeros' row and to reset indexes
    data = data.reset_index(drop=True)
    data = data.drop(0)
    data = data.reset_index(drop=True)
        
    return data
```

```python
vels = findVelocities(traj)
vels.to_pickle('velocities.pkl')
print(vels)
```

Estas velocidades están expresadas en píxeles/fotograma. Vamos a representar la trayectoria y velocidad en la direccion x de una partícula a lo largo de estos 80 fotogramas

```python
particle_2 = vels[vels.particle==2]

plt.scatter(particle_2.frame, particle_2.x)
plt.show()
plt.scatter(particle_2.frame, particle_2.vx)
plt.show()
```
